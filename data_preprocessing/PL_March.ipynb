```python
sku = df['Sku'].unique()
cat = df['Catalog'].unique()

sku_cat = {}

for s in sku:
    sku_cat[s] = {}
    for c in cat:
        sku_cat[s][c] = df[(df['Sku'] == s) & (df['Catalog'] == c)]['Catalog'].value_counts().sum()
        
```

```python
flare = {"name": "flare", "children": []}

for key, value in data.items():
    children = [{"name": k, "value": v} for k, v in value.items()]
    flare["children"].append({"name": key, "children": children})
```
